```python
copy['churn'].value_counts()/len(copy)*100
```

# numerical_columns

```python
numerical_columns=data[['credit_score','age','tenure','balance','products_number','estimated_salary']]
numerical_columns
```

# categorical_data

```python
categorical_data=data[['country','gender','churn','credit_card','active_member']]
```

```python
categorical_data
```

# Univariate analysis of numerical_columns

```python
# Univariate Analysis

plt.figure(figsize=(15, 10))
for idx, col in enumerate(numerical_columns):
    plt.subplot(3, 3, idx + 1)  
    sns.kdeplot(x=col, data=copy, hue="churn", fill=True, alpha=0.6)
    plt.title(f"Distribution of {col.capitalize()} by Churn")
    plt.xlabel(col.capitalize())
    plt.ylabel("Density")

plt.tight_layout()
plt.show()
```

# Univariate analysis of categorical_data

```python
plt.figure(figsize=(15, 10))
for idx, col in enumerate(categorical_data):
    plt.subplot(3, 3, idx + 1)  
    sns.countplot(x=col, data=copy, hue="churn", fill=True, alpha=0.6)
    plt.title(f"Distribution of {col.capitalize()} by Churn")
    plt.xlabel(col.capitalize())
    plt.ylabel("Density")

plt.tight_layout()
plt.show()
```

# BoxPlot

```python
# Boxplot
plt.figure(figsize=(15, 10))
for idx, col in enumerate(numerical_columns):  
    plt.subplot(3, 3, idx + 1)  
    sns.boxplot(x="churn", y=col, data=copy, palette="Set2")
    plt.title(f"Boxplot of {col.capitalize()} by Churn")
    plt.xlabel("Churn")
    plt.ylabel(col.capitalize())

plt.tight_layout()  # Prevent overlapping
plt.show()
```
